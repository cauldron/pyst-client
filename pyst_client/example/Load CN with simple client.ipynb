```python
from tqdm import tqdm
```

```python
import os
auth = os.environ["pyst_auth_token"]
assert auth

settings.api_key = auth
settings.server_url = "https://vocab.cauldron.ch"
```

```python
from pyst_client.cn import CombinedNomenclatureLoader
```

```python
cn = CombinedNomenclatureLoader(
    year=2024,
    api_key=auth,
    host="https://vocab.cauldron.ch/",
    sample=False
)
```

```python
self = cn
```

```python
from rdflib.namespace import SKOS
        
data = self.cn.expanded_separate_json_ld_graph(
    SKOS.broader,
    self.cn.relationships(kind=SKOS.broader, sample=self.sample),
)
```

```python
orjson.loads(data[0])
```

```python
from pyst_client.simple import *
```

```python
def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))
```

```python
for chunk in tqdm(chunker(data, 25)):
    selection = [orjson.loads(obj)[0] for obj in chunk]
    sources = [obj['@id'] for obj in selection]
    targets = [obj['http://www.w3.org/2004/02/skos/core#broader'][0]['@id'] for obj in selection]

    responses = Relationship.create_many(
        sources=sources,
        targets=targets,
        verbs=RelationshipVerbs.broader,
        timeout=240
    )
```
